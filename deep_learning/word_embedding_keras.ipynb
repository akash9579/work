```
from tensorflow.keras.preprocessing.text import one_hot
```

```
sent=[  'I support real madrid',
     'i play football',
     'real madrid win three UCL in roe',
    'romos was captain of real madrid',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```
voc_size=100
```

**one hot represenatation**

one_hot() is inbuild function of keras

we should pass word and voc_size

```
onehot_repr=[one_hot(words,voc_size)for words in sent] 
print(onehot_repr)
```

**Word Embedding Represntation**

for making word embedding matrix we have to give dimensions means no of features

```
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences    # all sentense should have same lenght sentense
from tensorflow.keras.models import Sequential
```

```
import numpy as np
```

```
sent_length=8      # max sentense length
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length) # one_hot_repe, pre padding applied in front of sentense
print(embedded_docs)
```

```
dim=10   # feature dimension
```

```
model=Sequential()    # making sequential model
model.add(Embedding(voc_size,10,input_length=sent_length)) # adding layers we have to give (voc size,dimension,max legth)
model.compile('adam','mse')
```

```
model.summary()
```

```
print(model.predict(embedded_docs))
```
