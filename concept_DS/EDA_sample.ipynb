**sns.heatmap**

```
import seaborn as sns                                                                             # sns.heatmap heatmap is A heat map is a data visualization technique that shows magnitude of a phenomenon as color in two dimensions
sns.heatmap(data.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

**sns.set_style** for setting the background as whitegrid

**sns.countplot** is for plotting 2d graph


```
sns.set_style('whitegrid')                                                                  # when ur column has less number of unique variable then use this one for data visualization
sns.countplot(x='Survived',data=data)
```

```
sns.set_style('whitegrid')
sns.countplot(x='Survived',hue='Sex',data=data,palette='RdBu_r')
```

```
sns.set_style('whitegrid')                                                                      # extra
sns.countplot(x='Survived',hue='Parch',data=data,palette='RdBu_r')
```

```
sns.set_style('whitegrid')
sns.countplot(x='Survived',hue='Pclass',data=data,palette='rainbow')
```

**correlation** **matrix**

**1 heatmap**

```
# calculate correlation matrix
corr = data.corr()# plot the heatmap 
```

```
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True)  
# most basic version
#1st we have to pass correlation matrix
# xticklabels,yticklabels basically ita for X,Y axis 
# annot interger value showed in box
```

```
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 100, as_cmap=True)) 
```
